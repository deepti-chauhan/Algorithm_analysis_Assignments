 Q-1: Suppose you are given a stack which can take only integers. As an input you are given
INPUT -> 10,20,30,40,50,60  inserted in stack from left to right. where,
    push-> A
    pop-> B
find order for 105040603020
HINT : order of 302050604010 is AAABBAABABBB

```python
#Solution 1:
#--------function---------#
def printOrder(myInput_array, myOrder_List):
    my_stack = []
    i=0
    while(len(myOrder_List)!=0):
        if(my_stack!=[] and (my_stack[-1] == myOrder_List[0])):
            my_stack.pop()
            print('B', end="")
            myOrder_List.pop(0)
        else:
            my_stack.append(myInput_array[i])
            print('A', end="")
            i+=1

#-----------main code-------------#
myInput_array = [10,20,30,40,50,60]
myOrder_List = [10,50,40,60,30,20]
print("\tThe Order will be : ", end='')
printOrder(myInput_array,myOrder_List)
```

Q-2:Given a stack of integers,how do you check whether each successive pair of numbers in the stack is consecutive or not.
The pair can be increasing or decreasing and is stack has odd numbers of elements, the element at the top is left out of the pair.
INPUT-> [4,5,-2,-3,11,10,5,6,20]
OUTPUT->(4,5), (-2,-3), (11,10), (5,6)

```python
#Solution 2:
def find_pairs(stk1):
    for i in range(len(stk1)-1):
        if(abs(stk1[i] - stk1[i+1])==1):
            print('('+str(stk1[i]), str(stk1[i+1])+')', end =", ")

stk1 = [4,5,-2,-3,11,10,5,6,20]
print("Given stack : ", str(stk1))
print("Consecutive pairs in given stack : ")
find_pairs(stk1)
```

Q-3: Consider a ticket line at reservation counter. Ticket distributor employee is crazy and he 
wants reverse the order of the first m persons from the queue. others will be stand at their place,
for example, m=4 and line has 9 members so 
INPUT ->    [10 20 30 40 50 60 70 80 90]
OUTPUT ->   [40 30 20 10 50 60 70 80 90]

```python
#Solution 3:

#------function------#
def reverse_order(arr, m):
    array_length = len(arr)
    if(m==0):
        print("Same order for m=0 : "+str(arr))
    elif(m <= array_length):
        reverse_section = arr[m-1::-1]
        remaining_array = arr[m:]
        reverse_section.extend(remaining_array)
        print("Reverse Queue : "+str(reverse_section))
    else: 
        print("\t$-----------m = out of index range-----------$")

#-----------main code-----------#
input_array = [10,20,30,40,50,60,70,80,90] 
m = int(input()) #m = 4

print("Given token id in a queue : " + str(input_array))
reverse_order(input_array, m)
```
